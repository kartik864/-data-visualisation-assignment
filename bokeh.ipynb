```python
#1
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

# Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, line_width=2)

show(p)
```

```python
#2
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

output_notebook()

# Data
np.random.seed(25)
data = {
    'x': np.random.rand(50),
    'y': np.random.rand(50),
    'sizes': np.random.randint(10, 100, 50),  # Bubble sizes
    'colors': np.random.rand(50)  # Colors
}
df = pd.DataFrame(data)

# Plot
p = figure(title="Scatter Plot with Bubble Sizes and Colors", x_axis_label='x', y_axis_label='y')
p.scatter('x', 'y', size='sizes', color='colors', source=df, alpha=0.6)

show(p)
```

```python
#3
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import pandas as pd

output_notebook()

# Updated Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create DataFrame
data = {
    'fruit': fruits,
    'count': counts
}
df = pd.DataFrame(data)

# Plot
p = figure(x_range=df['fruit'], plot_height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x='fruit', top='count', width=0.9, source=df, legend_field="fruit",
       line_color='white', fill_color='green')

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = 'Fruit'
p.yaxis.axis_label = 'Count'
p.xaxis.major_label_orientation = 1
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

show(p)
```

```python
#4
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

output_notebook()

# Data
np.random.seed(30)
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Prepare data for Bokeh
hist_df = pd.DataFrame({
    'left': edges[:-1],
    'right': edges[1:],
    'count': hist
})

# Plot
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency', 
           plot_height=350, plot_width=700)

p.quad(top='count', bottom=0, left='left', right='right', 
        source=hist_df, fill_color='navy', line_color='white', alpha=0.7)

show(p)
```
