```python
#1
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import pandas as pd

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X'], df['Y'], df['Z'], c='b', marker='o')

ax.set_title('3D Scatter Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
#2
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.violinplot(x='Grade', y='Score', data=df)
plt.title('Violin Plot of Scores Across Grades')
plt.xlabel('Grade')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

```python
#3
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Seed and generate sales data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data for heatmap
sales_pivot = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(sales_pivot, cmap='YlGnBu', annot=True, fmt='.0f', linewidths=0.5)
plt.title('Heatmap of Sales Across Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

```python
#4
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis')

ax.set_title('3D Surface Plot')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()



```

```python
#5
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Seed and generate data
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], 
            s=df['Population'] * 0.5,  # Bubble size proportional to population
            alpha=0.7, 
            c='b', 
            edgecolors='w', 
            linewidth=2)

# Adding annotations
for i, row in df.iterrows():
    plt.text(row['GDP'], row['Population'], row['Country'], 
             fontsize=12, ha='right')

plt.title('Bubble Chart: GDP vs Population')
plt.xlabel('GDP')
plt.ylabel('Population')
plt.grid(True)
plt.show()




```
